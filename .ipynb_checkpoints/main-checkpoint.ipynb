```python
dataset_paths = dataset.load_dataset_path()
```

```python
def load_text(path):
    with open(path) as f:
        s = f.read()
        s = s.replace("\n", " ")
        s = s.replace("\'", "")
        s = s.replace("\"", "")
        s = s.replace(".", "")
        s = s.replace(",", "")
        s = s.replace("?", "")
        s = s.replace(":", "")
        s = s.replace("  ", " ")
        s = s.strip()
        return s

s = load_text("./dataset/articles/entertainment/289.txt")
s
```

```python
from tensorflow.keras.layers import TextVectorization

max_tokens = 100

vectorize_layer = TextVectorization(
    max_tokens=max_tokens,
    split='character',
    output_mode='int',
    output_sequence_length=1)

for [article_path, summary_path] in dataset_paths:
    article = dataset.load_text(article_path)
    summary = dataset.load_text(summary_path)
    vectorize_layer.adapt([article, summary])
```

```python
for [article_path, summary_path] in dataset_paths[:1]:
    summary = dataset.load_text(summary_path)
    vector = vectorize_layer(list(summary))
    print(vector)
```

```python
EOS = '[EOS]'
BOS = '[BOS]'
UNK = '[UNK]'

def get_token(char):
    if char == EOS:
        return 0
    
    if char == BOS:
        return 1

    if char == ' ':
        return 2

    char_code = ord(char)

    if char_code < 65 or char_code > 122:
        return 3 #Unkown char
    
    return char_code  - 61

def get_text(token):
    if token == 0:
        return EOS
        
    if token == 1:
        return BOS

    if token == 2:
        return ' '

    if token < 4 or token > 61:
        return UNK

    return chr(token + 61)

def vetorize_sentence(sen):
    vector = [[get_token(BOS)]]
```
